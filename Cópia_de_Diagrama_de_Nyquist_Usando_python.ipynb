<h1>\begin{align}G(s) = \frac{1}{s-0.5}

```python
from pylab import plot
import control
from control import TransferFunction

s = TransferFunction.s
k = 1
G  = (1)/((s-0.5))

control.nyquist(G)

plot()
```
